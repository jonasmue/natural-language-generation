## Text Generation

```
from transformers import AutoTokenizer, GPT2LMHeadModel
```

```
TEXT_START = "Human Computer Interaction, Deep Learning, Machine Learning, Psychology, User Experience, Natural Language Processing, Software Engineering\nDear Mr. Morley,"
```

```
MODEL_PATH = "/content/drive/My Drive/data/cover_letter_model"
```

```
tokenizer = AutoTokenizer.from_pretrained(MODEL_PATH)
model = GPT2LMHeadModel.from_pretrained(MODEL_PATH)
```

```
input_ids = tokenizer.encode(TEXT_START, return_tensors="pt")
```

```
sample_outputs = model.generate(
    input_ids,
    do_sample=True,
    max_length=512,
    top_k= 50,
    top_p=0.92,
    num_return_sequences=3
)
```

```
print("Output\n" + "=" * 80)
for i, sample_output in enumerate(sample_outputs):
  print("{}: {}".format(i, tokenizer.decode(sample_output, skip_special_tokens=True)))
```
